# Job Market Data Cleaning Report

```python
import pandas as pd
import glob

# Get all NON-checkpoint CSV files (final outputs only)
csv_files = [f for f in glob.glob("*.csv") if not f.startswith("checkpoint")]

print("Found CSV files:", csv_files)

dfs = [pd.read_csv(f) for f in csv_files]

# Combine
full_df = pd.concat(dfs, ignore_index=True)
print("Total rows before dedup:", len(full_df))
```

```python
full_df = full_df.drop_duplicates(subset=["Links"])
print("Total rows after deduplication:", len(full_df))
```

```python
clean_df = full_df.copy()

# Remove missing titles or companies
clean_df = clean_df.dropna(subset=["Title", "Company"])

# Remove bad or failed descriptions
clean_df = clean_df[~clean_df["Description"].str.contains("Error retrieving", na=False)]

print("Final cleaned rows:", len(clean_df))
```
